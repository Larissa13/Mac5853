```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
```

```python
Seen = []

def text_from_web(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
    req = Request(url=url, headers=headers) 
    global Seen
    Seen += [url]
    html = urlopen(req).read()
    soup = BeautifulSoup(html, "lxml")
    links = []
    for script in soup(["script", "style"]):
        script.extract()

    text = soup.body.get_text()

    lines = (line.strip() for line in text.splitlines())

    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))

    text = '\n'.join(chunk for chunk in chunks if chunk)
    
    return text
```

```python
#TODO fzer um crawler q entra em todos os links do site ateh certa profundidade e extrai o texto em todos

def crawler(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
    req = Request(url=url, headers=headers) 
    html = urlopen(req).read()
    soup = BeautifulSoup(html, "lxml")
    possible_links = soup.find_all('a', href=True)
    links = []
    for link in possible_links:
        link_url = link.get('href')
        if link_url[0] == '/':
            links += [url[:-1] + link_url]
        elif 'https' in link_url:
            links += [link_url]
        else:
            links += [url + link_url]
    return links
```

```python
class TestTextExtraction(ut.TestCase):
    simple_url = "https://linux.ime.usp.br/~larissadeop/"
    complex_url = "https://linux.ime.usp.br/~larissadeop/Teste_mac5853/"
    html_content = "This is a test page.\nLet\'s find out if the code can detect these texts.\nNow a link"
    crawler_url = "https://linux.ime.usp.br/~larissadeop/Mac0499/"
    links = ['https://linux.ime.usp.br/~larissadeop/Mac0499/Proposta TCC.pdf', 
             'https://linux.ime.usp.br/~larissadeop/Mac0499/Monografia.pdf', 
             'https://linux.ime.usp.br/~larissadeop/Mac0499/apreciacao.pdf', 
             'https://linux.ime.usp.br/~larissadeop/Mac0499/poster.pdf', 
             'https://linux.ime.usp.br/~larissadeop/Mac0499/Apresentacao.pdf']
    
    def test_simple_text_from_web(self):
        self.assertEqual(text_from_web(self.simple_url), '>\nMac0499\nTesteMac5853')
        
    def test_complex_text_from_web(self):
        self.assertEqual(text_from_web(self.complex_url), self.html_content)
    
    def test_crawler(self):
        self.assertEqual(crawler(self.complex_url), ['https://www.google.com'])
    
    def test_crawler_complex(self):
        self.assertEqual(crawler(self.crawler_url), self.links)
    
    
# if __name__ == '__main__':
#     ut.main() only for script
if __name__ == '__main__':
    ut.main(argv=['first-arg-is-ignored'], exit=False)
    
```
