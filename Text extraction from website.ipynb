```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
```

```python
Seen = []

def text_from_web(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
    req = Request(url=url, headers=headers) 
    global Seen
    Seen += [url]
    html = urlopen(req).read()
    soup = BeautifulSoup(html)
    links = []
    for script in soup(["script", "style"]):
        script.extract()

    text = soup.body.get_text()

    lines = (line.strip() for line in text.splitlines())

    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))

    text = '\n'.join(chunk for chunk in chunks if chunk)
    
    return text
```

```python
#TODO fzer um crawler q entra em todos os links do site ateh certa profundidade e extrai o texto em todos

def crawler(url):
    print("URL", url)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
    req = Request(url=url, headers=headers) 
    html = urlopen(req).read()
    soup = BeautifulSoup(html)
    possible_links = soup.find_all('a', href=True)
    links = []
    for link in possible_links:
        if link.attrs['href'][0] == '/':
            links += [url[:-1] + link.attrs['href']]
        else:
            links += [link.attrs['href']]
    return links
```
